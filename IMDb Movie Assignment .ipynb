-  ### Subtask 2.3: Drop unecessary rows using columns with high Null percentages

Now, on inspection you might notice that some columns have large percentage (greater than 5%) of Null values. Drop all the rows which have Null values for such columns.

```python
# Write your code for dropping the rows here

'''
drop_data is a series of left out column names with null percentages after 
dropping several unnecessary columns from the movies dataframe.

drop_dict is the variable that holds the above series but in dictionary format.

drop_list is a list of columns filtered when looking for columns which has 
null percentage above 5.00%. We use for loop to iterate through drop_dict 
dictionary to filter out and contain the column names in this this.

Finally we drop all the rows at the end with all NA with respect to columns 
obtained in the last step. We pass the drop_list list to subset parameter
to achieve the same.
'''

drop_data=round(100*(movies.isnull().sum(axis = 0)/len(movies)),2)
#print(drop_data)
drop_dict = drop_data.to_dict()
#print(drop_dict)
drop_list=[]
for column, na in drop_dict.items():
    if na > 5.00:
        drop_list.append(column)
#print(drop_list)
movies = movies.dropna(axis=0, subset=drop_list)
movies
```

Create three new dataframes namely, `Meryl_Streep`, `Leo_Caprio`, and `Brad_Pitt` which contain the movies in which the actors: 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' are the lead actors. Use only the `actor_1_name` column for extraction. Also, make sure that you use the names 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' for the said extraction.
    2. Append the rows of all these dataframes and store them in a new dataframe named `Combined`.
    3. Group the combined dataframe using the `actor_1_name` column.

For example, the  `title_year`  year 1923, 1925 should be stored as 1920s. Sort the dataframe based on the column `decade`, group it by `decade` and find the sum of users voted in each decade. Store this in a new data frame called `df_by_decade`.

```python
# Write your code for creating three new dataframes here
# Include all movies in which Meryl_Streep is the lead

Meryl_Streep = movies[movies['actor_1_name'] == 'Meryl Streep']
Meryl_Streep[['movie_title', 'actor_1_name']]
```

```python
Combined.isnull().sum(axis = 0)
```

```python
Combined['num_critic_for_reviews'].isnull().sum()
```

```python
# Write the code for finding the mean of critic reviews and audience reviews here
critic = Combined['num_critic_for_reviews'].mean()
critic = critic.sort_values(by='num_critic_for_reviews', ascending=False).head()
#critic

audience = Combined['num_user_for_reviews'].mean()
audience = audience.sort_values(by='num_user_for_reviews', ascending=False).head()
#audience

reviews = pd.concat([critic, audience], axis=1)
reviews
```

```python
# Write the code for calculating decade here

movies.loc[(movies['title_year'] < 1930) & (movies['title_year'] >= 1920), 'decade'] = '1920s'
movies.loc[(movies['title_year'] < 1940) & (movies['title_year'] >= 1930), 'decade'] = '1930s'
movies.loc[(movies['title_year'] < 1950) & (movies['title_year'] >= 1940), 'decade'] = '1940s'
movies.loc[(movies['title_year'] < 1960) & (movies['title_year'] >= 1950), 'decade'] = '1950s'
movies.loc[(movies['title_year'] < 1970) & (movies['title_year'] >= 1960), 'decade'] = '1960s'
movies.loc[(movies['title_year'] < 1980) & (movies['title_year'] >= 1970), 'decade'] = '1970s'
movies.loc[(movies['title_year'] < 1990) & (movies['title_year'] >= 1980), 'decade'] = '1980s'
movies.loc[(movies['title_year'] < 2000) & (movies['title_year'] >= 1990), 'decade'] = '1990s'
movies.loc[(movies['title_year'] < 2010) & (movies['title_year'] >= 2000), 'decade'] = '2000s'
movies.loc[(movies['title_year'] < 2020) & (movies['title_year'] >= 2010), 'decade'] = '2010s'
movies[['movie_title', 'title_year', 'decade']]
```

```python
# Write your code for creating the data frame df_by_decade here 

df_by_decade = movies.sort_values(by='decade').groupby('decade', as_index=False)['num_voted_users'].agg('sum')
df_by_decade
```

```python
# Write your code for plotting number of voted users vs decade

X=df_by_decade['decade']
Y=df_by_decade['num_voted_users']

plt.figure(figsize=(20,6))
plt.rcParams['font.size'] = 14

bars = plt.bar(X,Y, color='deepskyblue', width=0.50, log=True)

bars[df_by_decade['num_voted_users'].idxmin()].set_color('red')
bars[df_by_decade['num_voted_users'].idxmax()].set_color('lime')

plt.title('No. of Voted Users each Decade', y=1.0, fontsize='24')
plt.xlabel('Decade', fontsize=24, fontname="Comic Sans MS")
plt.ylabel('No. of Voted Users', fontsize=24, fontname="Comic Sans MS" )

plt.grid()


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval, yval, fontsize=15)


plt.show()
```

<div class="alert alert-block alert-info">
<b>This bar chart represents each decade with no. of voted users on the top where 1920s had the least no. shown in red color and 2000s had the highest one shown in green color</b>   
</div>
